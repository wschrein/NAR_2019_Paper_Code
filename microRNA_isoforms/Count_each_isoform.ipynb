```python
def readfastq(filename):
    sequences=[]
    with open(filename) as fh:
        while True:
            fh.readline()
            seq=fh.readline().rstrip()
            fh.readline().rstrip()
            fh.readline()
            if len(seq)==0:
                break
            sequences.append(seq)
        return sequences
    
```
